```python
import pandas as pd
import re
```

```python
with open('WhatsApp Chat with Announcements BSCS 7(A).txt','r',encoding= 'utf-8') as f:
      data= f.read()
      print(data)
```

```python
#pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s' 
x = re.split(pattern, data)[1:]
messages = re.split(pattern, data)[1:]
dates = re.findall(pattern, data)
df = pd.DataFrame({'message_date': dates,'user_message': messages})
```

```python
df
```

```python
date = []
time= []
for i in df.message_date:
    date.append(i.split(',')[0])
    time.append(i.split(',')[1])
    
df['date']= date   
df['time']= time   
```

```python
#For i in df['message_date']:
#    print(i)#df['message_date'].split(',')
df    
```

```python
df['time']= df['time'].apply(lambda x : x.replace('-',''))
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['user'] = users
df['message'] = messages
df.drop('user_message',axis='columns',inplace=True)

```

```python
s ='group_notification'
len((df[df['user']==s]).message)
```

```python
members = df.user.unique()
```

```python
df['date'].convert_dtypes(convert_string=False)
df['time'].convert_dtypes(convert_string=False)
#a= pd.to_datetime(df['time'], format='%H:%M').dt.normalize()
```

```python
#pd.to_datetime(df['message_date'], format='%d/%m/%Y, %H:%M - ')
df['date']= pd.to_datetime(df['message_date'], format='%m/%d/%y, %H:%M - ')
```

```python
df['only_date'] = df['date'].dt.date
df['year'] = df['date'].dt.year
df['month_num'] = df['date'].dt.month
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['day_name'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```

```python
df.sample(10)
```

```python
period = []
for hour in df[['day_name', 'hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour + 1))
    else:
        period.append(str(hour) + "-" + str(hour + 1))

df['period'] = period

df.sample(5)
```
